```python
import pandas as pd
import matplotlib.pyplot as plt
from database.spotify import Spotify
```

```python
spotify = Spotify()
```

```python
spotify.connect()
report = spotify.retrieve("created_lists")
spotify.disconnect()
```

```python
analysis = report.groupby(["seed_song","follower_based"]).mean().sort_values("playlist_size",ascending=False).reset_index()
analysis["score"] = (analysis["playlist_size"]/analysis["playlist_size"].max()) * (analysis["count"] / analysis["count"].max())
a =analysis.sort_values("score",ascending=False).head(50)
```

```python
a
```

```python
for follower in [True,False]:
    aggregates = []
    for i in range(50):
        try:
            data = a.iloc[i]
            song_data = report[(report["seed_song"]==data["seed_song"]) & (report["follower_based"]==follower)]
            song_data.reset_index(drop=True,inplace=True)
            song_data["key"] = song_data.index
            aggregates.append(song_data)
        except:
            continue
    visuals = pd.concat(aggregates).groupby("key").mean()
    plt.plot(visuals["count"])
    plt.show()
```
